```python
import pandas as pd
import json

# Dosya isimlerinin listesi
filenames = ["register_0025.csv","register_0062.csv", "register_0060.csv", "register_0050.csv"]

word_counts_per_file = {}
total_word_counts = {}

for filename in filenames:
    data = pd.read_csv(path + filename)

    word_counts = {}

    for index, row in data.iterrows():
        region_attributes = json.loads(row['region_attributes'])

        if 'word' in region_attributes:
            word = region_attributes['word']
            word_counts[word] = word_counts.get(word, 0) + 1
            total_word_counts[word] = total_word_counts.get(word, 0) + 1

    word_counts_per_file[filename] = word_counts
```

## Genel Kelime Istatistikleri

```python
for filename, counts in word_counts_per_file.items():
    print(f"Istatistikler {filename} için:")
    for word, count in counts.items():
        print(f"  {word}: {count}")
    print()

print("Tüm dosyalar için toplam istatistikler:")
for word, count in total_word_counts.items():
    print(f"  {word}: {count}")
```

### Histogram

```python
import pandas as pd
import json
import matplotlib.pyplot as plt


for filename, counts in word_counts_per_file.items():
    plt.figure(figsize=(10, 5))
    plt.bar(counts.keys(), counts.values())
    plt.title(f"Istatistikler {filename} için")
    plt.xlabel('Kelime')
    plt.ylabel('Frekans')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

plt.figure(figsize=(10, 5))
plt.bar(total_word_counts.keys(), total_word_counts.values())
plt.title("Tüm dosyalar için toplam istatistikler")
plt.xlabel('Kelime')
plt.ylabel('Frekans')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

## Belge Basina Istatistikler
#### 1- Toplam belge sayisi ve toplam kelime sayisi
#### 2- Her belgede ortalama gecen kelime sayisi
#### 3- Her bir kelimenin herbir belgede ortalama kac defa gectigi
#### 4- Her bir dosyadaki belge basina dusen ortalama kelime sayisi



```python

total_documents = 0
total_words = sum(total_word_counts.values())

document_counts = {}
for filename in filenames:
    data = pd.read_csv(path+filename)
    document_counts[filename] = len(data['filename'].unique())
    total_documents += document_counts[filename]

average_words_per_document = total_words / total_documents

average_occurrences_per_word_per_document = {
    word: count / total_documents for word, count in total_word_counts.items()
}

print(f"Toplam belge sayısı: {total_documents}")
print(f"Toplam kelime sayısı: {total_words}")
print(f"Belge başına ortalama kelime sayısı: {average_words_per_document:.2f}")
print("\nHer kelimenin belge başına ortalama geçme sayıları:")
for word, avg in average_occurrences_per_word_per_document.items():
    print(f"  {word}: {avg:.2f}")

print()
average_words_per_document_per_file = {}
for filename, counts in word_counts_per_file.items():
    average_words_per_document_per_file[filename] = sum(counts.values()) / document_counts[filename]
    print(f"{filename} içindeki belge başına ortalama kelime sayısı: {average_words_per_document_per_file[filename]:.2f}")
```

#### Hangi belgede kactane kelime var

```python
import pandas as pd

words_per_document = {}

filenames = ["register_0025.csv", "register_0062.csv", "register_0060.csv", "register_0050.csv"]

for filename in filenames:
    data = pd.read_csv(path + filename)
    for doc in data['filename'].unique():
        words_per_document[doc] = words_per_document.get(doc, 0) + len(data[data['filename'] == doc])

sorted_words_per_document = sorted(words_per_document.items(), key=lambda item: item[1], reverse=True)

for doc, count in sorted_words_per_document:
    print(f"'{doc}' belgesinde {count} tane kelime var.")
```
